# Задание 1
Скачайте с сайта grouplens.org...movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd

data = pd.read_csv('ml-latest-small/movies.csv') #movieId  title
data2 = pd.read_csv('ml-latest-small/ratings.csv') #userId movieId rating 

data2 = data2[data2['rating']== 5.0].groupby('movieId').count().sort_values(by='rating',ascending=False)

joined = data2.merge(data, on='movieId', how = 'left')
joined[['title','rating']].head()
```
