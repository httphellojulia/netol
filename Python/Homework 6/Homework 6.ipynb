```python
import json

final_dict = {}

with open('downloads/purchase_log.txt', 'r') as f:
    for i,line in enumerate(f):
        if i>1680 and i < 1750:
            dict_ = json.loads(line)
            key = dict_['user_id']
            values = dict_['category']
            if key not in final_dict.keys():
                final_dict.setdefault(key,[values])
            else:
                final_dict[key].append(values)
                print('Добавление работает',final_dict[key],'\n' )
print(final_dict)
```

```python
import json

with open('downloads/visit_log.csv', 'r') as f2read:
    with open('downloads/funnel.csv', 'w') as f2write:
        for i, line in enumerate(f2read):
            if i > 1:
                list_ = line.strip().split(',')
                if list_[0] in final_dict.keys():
                    #print("Совпадение найдено")
                    list_.append(str(final_dict[list_[0]]).strip("'']["))
                    f2write.write(str(list_))
    f2write.close()
    f2read.close()
    
with open('downloads/funnel.csv', 'r') as f2write:
    print(f2write.readline())
```
