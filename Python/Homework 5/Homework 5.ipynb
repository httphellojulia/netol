```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
         
def p(number):
    for i in documents:
        if number == i['number']:
            return print('Владелец документа:',i['name'])
    print('Документ не найден в базе')

def s(number):
    if polki(number) != None:
        print('Документ хранится на полке: ', polki(number))
    else:
        print('Документ не найден в базе')
    
def polki(number):
    for key, value in directories.items():
        for i in value:
            if number == i:
                return key
def l():
    for i in documents:
        print('№:', i['number'],', тип:', i['type'], ', владелец:',i['name'], ', полка хранения:', polki(i['number']))
        
def current_polki():
    print('Текущий перечень полок:', ', '.join(directories.keys()))

def ads(number):
    if number in directories.keys():
        print('Такая полка уже существует.')
    else:
        directories.update({number:''}) 
        print('Полка добавлена.')
    current_polki()
    
def ds(number):
    if number not in directories.keys():
        print('Такой полки не существует.')
    elif bool(directories[number]) == 1:
        print('На полке есть документы, удалите их перед удалением полки.')
    else:
        directories.pop(number)
        print('Полка удалена.')
    current_polki()

def start(command):
    while command != 'q':
        if command == 'p':
            p(input("Введите номер документа "))
        elif command == 's':
            s(input("Введите номер документа "))
        elif command == 'l':
            l()
        elif command== 'ads':
            ads(input("Введите номер полки:"))
        elif command== 'ds':
            ds(input("Введите номер полки:"))
        else:
            print('Такой команды нет. Доступные команды:\np - поиск владельца по номеру документа\ns - поиск полки на которой находится документ\nl - полная информация по документам\nads - добавление полки\nds - удаление полки\nq - завершение программы')
        command = input("Введите номер команды ")
        
start(input("Введите номер команды "))
```
